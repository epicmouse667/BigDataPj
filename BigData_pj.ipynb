```python
import pandas as pd
names=["ICAO","IATA","name","city","country","latitude_deg","latitude_min","latitude,sec",
"latitude_dir","longitude_deg","longitude_min","longitude_sec","longitude_dir","altitude","latitude","longitude"]
airport_df=pd.read_table("/content/GlobalAirportDatabase.txt",sep=":",header=None,names=names)
airport_df
```

# Kmeans

```python
from sklearn.cluster import KMeans
import numpy as np
latitude_array=airport_df["latitude"].to_numpy(dtype="float").reshape(-1,1)
longitude_array=airport_df["longitude"].to_numpy(dtype="float").reshape(-1,1)
X=np.concatenate((latitude_array,longitude_array),axis=1)
kmeans = KMeans(n_clusters=12, random_state=0).fit(X)
kmeans.cluster_centers_
```

# throughput dataset

```python
import pandas as pd
throughput_df=pd.read_excel("/content/drive/MyDrive/bigdata_pj/airport_throughput.xlsx")
throughput_df.rename(columns={"Airport":"name"},inplace=True)
throughput_df.dropna()
```

```python
df=pd.merge(airport_df[["ICAO","latitude","longitude"]],throughput_df[["name","ICAO","Enplanements"]])
df=df[df["latitude"]!=0.000]
df=df.reset_index()
df=df.drop(["index"],axis=1)
df
```

```python
import overpy
import pandas as pd
import time
from tqdm import tqdm
# import geopandas as gpd
api = overpy.Overpass()
latitude=df["latitude"].to_numpy()
longitude=df["longitude"].to_numpy()
features=["natural","amenity","shop","railway","public_transport"]
for i in tqdm(range(latitude.shape[0])):
  for feature in features:
    # avoid too frequent request
    time.sleep(2)
    query_result = api.query(f"""
    [out:json];
    node(around:10000.0,{latitude[i]},{longitude[i]})[name][{feature}];
    out body;""")
    df.at[i,feature]=len(query_result.nodes)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
df=pd.read_csv("/content/drive/MyDrive/bigdata_pj/all_poi.csv")
X=df.drop(["Enplanements","ICAO","name"],axis=1).values
y=df["Enplanements"].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
import tensorflow as tf
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(16,activation="tanh", input_shape=(X_train.shape[1],)))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dense(32,activation="tanh"))
model.add(tf.keras.layers.Dense(32,activation="relu"))
model.add(tf.keras.layers.Dense(8,activation="relu"))
model.add(tf.keras.layers.Dense(1,activation="relu"))
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), loss='mse')
model.fit(X_train, y_train, batch_size=32, epochs=20)
```
